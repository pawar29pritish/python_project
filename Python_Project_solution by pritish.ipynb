```python
# import all required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define schema
customer_columns_list = ['custno','firstname','lastname','gender','age','profession','contactNo','emailId','city','state','isActive','createdDate','updatedDate']
product_columns_list = ['productno','productName','Description','isActive','createdDate','updatedDate']
transaction_columns_list = ['txnno','txndate','custno','Amount','productno','spendby']
```

```python
# load the data in DataFrame
df_customers_s1_csv = pd.read_csv('customers.csv',names = customer_columns_list)
# show the top 5 record of the data 
df_customers_s1_csv.head()
```

```python
df_products_s1_csv = pd.read_csv('products.csv',names = product_columns_list)
df_products_s1_csv.head()
```

```python
df_transactions_s1_csv = pd.read_csv('transactions.csv',names = transaction_columns_list)
df_transactions_s1_csv.head()
```

```python
# Usecase-2 Display only those customers from CSV_s1, who purchased more than 3 products
df_transactions_s1_csv.groupby('custno').filter(lambda df_transactions_s1_csv : df_transactions_s1_csv['custno'].count() > 3)['custno'].unique()
```

```python
# Usecase-3 Display top 5 most demanded products from CSV_s1
df_transactions_s1_csv['productno'].value_counts().head()
```

```python
# Usecase-4 Display top 5 transactions amount from CSV_s1
df_transactions_s1_csv.sort_values('Amount',ascending=False).head()['Amount']
```

```python
# Usecase-5 Display distinct professions from CSV_s1
df_customers_s1_csv['profession'].unique()
```

```python
# Usecase-6 Display highest age in CSV_s1 customer’s dataset
df_customers_s1_csv['age'].max()
```

```python
# Usecase-7 Display custno, gender, age, profession, contactno, productno, productName, txndate, spendby, amount from CSV_s1 for custno = 923
df1 = pd.merge(df_transactions_s1_csv[df_transactions_s1_csv['custno'] == 923] , df_products_s1_csv , on='productno',how='inner')
```
